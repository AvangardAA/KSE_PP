```python
from collections import Counter
global word, relfreq
freq = ()
with open("output.txt", "r") as f:
    c = f.read()
    freq = Counter(c)
    word = c

encrsum = sum(freq.values())

relfreq = {char: count / encrsum for char, count in freq.items()}

print(relfreq)
```

```python
diff = []
for key in range(26):
    encr_out = ""
    for c in word:
        if c.isalpha():
            cInUp = c.isupper()
            c = c.lower()
            encC = chr(((ord(c) - ord('a') - key) % 26) + ord('a'))
            if cInUp:
                encC = encC.upper()
            encr_out += encC
        else:
            encr_out += c
    summ = 0
    decrCounter = Counter(encr_out)
    decrsum = sum(decrCounter.values())
    relfreqdecr = {char: count / decrsum for char, count in decrCounter.items()}
    diff.append({key: sum(reffreq[keyy] - relfreqdecr[keyy] for keyy in relfreqdecr)})
```
